### My solution

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 20001)

y1 = 0.0                             # x < -2.0
y2 = x + 2.0                         # -2.0 ≤ x < 1.0
y3 = 4.0 - x                         # 1.0 ≤ x < 4.0
y4 = 0.0                             # x ≥ 4.0

#no if-statements
m1 = (x <  -2.0)
m2 = (x >= -2.0) & (x <  1.0)
m3 = (x >=  1.0) & (x <  4.0)
m4 = (x >=  4.0)

#Combine masks with expressions
y = m1*y1 + m2*y2 + m3*y3 + m4*y4

plt.figure(figsize=(9,5))
plt.plot(x, y, lw=2.0, label="f(x)")
plt.title("Problem 1: Piecewise Function (triangle shape)")
plt.xlabel("x"); plt.ylabel("f(x)")
plt.grid(True, alpha=0.35); plt.legend()
plt.show()
```

### My solution

```python
x = np.linspace(-10, 10, 20001)

g1 = 0*x - 880.94
g2 = (1.0*x**3) - (12.05*x**2) + (39.58*x) + 22.58
g3 = 0*x + 10.97

#Boolean masks
n1 = (x <  -5.9)
n2 = (x >= -5.9) & (x <  7.2)
n3 = (x >=  7.2)

#Combine
g = n1*g1 + n2*g2 + n3*g3

plt.figure(figsize=(9,5))
plt.plot(x, g, lw=2.0, label="f(x)")
plt.title("Problem 2: Piecewise Function (boolean masks)")
plt.xlabel("x"); plt.ylabel("f(x)")
plt.grid(True, alpha=0.35); plt.legend()
plt.show()
```
