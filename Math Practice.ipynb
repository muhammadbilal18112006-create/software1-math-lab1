```python
#chapter 2

import numpy as np
import matplotlib.pyplot as plt
```

```python
#create 4th polynomial function

x = np.linspace(0,4)
y = 0.1 * x**4 - 0.5 * x**3 + 0.2 * x**2 - 0.1 * x +4

fig, ax = plt.subplots(figsize=(10,7))
plt.plot(x,y, linestyle='__', linewidth=5, label='4th polynomial function')
plt.grid(True)
plt.xlabel('x', fontsize = 15)
plt.ylabel('y', fontsize = 15)
plt.title('graph of function', fontsize=24)
plt.legend()
plt.show()
```

```python
fig1, ax1 = plt.subplots() #create a figure including an axes
x = np.linspace(0,4)
y1 = x -3

plt.plot(x,y1, color = 'y', label = 'x - 3')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
#plt.show()


#for y2
y2 = x**2 - 4*x
fig2, ax2 = plt.subplots()
plt.plot(x, y2, color = 'b', label = '$x^2 - 4x$')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
#ploting two graphs on single page

fig, ax = plt.subplots(2, figsize = (10,8))
x = np.linspace(0,4)
y1 = x -3
plt.sca(ax[0])
plt.plot(x,y1, color = 'y', label = 'x - 3')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

y2 = x**2 - 4*x
plt.sca(ax[1])
plt.plot(x, y2, color = 'b', label = '$x^2 - 4x$')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
