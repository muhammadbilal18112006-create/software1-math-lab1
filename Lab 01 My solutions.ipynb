## My solution for problem 2

```python
# Problem 2 - finding intersection points

# overview with y=1 line
plt.figure()
plt.plot(x, y)
plt.axhline(y=1, color='r', linestyle='--')
plt.title("f(x) with y=1 line for student 2601860")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

```python
# zoomed view - first crossing point

plt.figure(figsize=(7, 5))
plt.plot(x, y)
plt.xlim(-8.5, -7.5)
plt.ylim(0, 2)
plt.axhline(y=1, color='r', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Intersection point 1 - zoomed view')
plt.grid(True)
plt.show()
```

```python
# zoomed view - second crossing point

plt.figure(figsize=(7, 5))
plt.plot(x, y)
plt.xlim(3.0, 5.0)
plt.ylim(0, 2)
plt.axhline(y=1, color='r', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Intersection point 2 - zoomed view')
plt.grid(True)
plt.show()
```
